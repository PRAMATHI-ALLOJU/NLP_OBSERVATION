```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')  # Download required data once

lemmatizer = WordNetLemmatizer()

words = ['running', 'cats', 'better', 'wolves', 'was']

# Convert words to their base form
lemmas = [lemmatizer.lemmatize(word) for word in words]

print("Original:", words)
print("Lemmatized:", lemmas)
```

```python
import nltk
from nltk.stem import PorterStemmer
nltk.download('punkt')

text = "The striped bats are hanging."
words = nltk.word_tokenize(text)

stemmer = PorterStemmer()
stems = [stemmer.stem(w) for w in words]

print("Original Words:", words)
print("Stemmed Words:", stems)
```

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger') # Changed to download the correct resource

text = "The children are playing."
words = nltk.word_tokenize(text)

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def pos(w):
    # Removed explicit tagset specification
    tag = nltk.pos_tag([w])[0][1][0].upper()
    return {'J': wordnet.ADJ,'N': wordnet.NOUN,'V': wordnet.VERB,'R': wordnet.ADV}.get(tag, wordnet.NOUN)

for w in words:
    print(f"{w:10} → Stem: {stemmer.stem(w)}, Lemma: {lemmatizer.lemmatize(w, pos(w))}")
```
